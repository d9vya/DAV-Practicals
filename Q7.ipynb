# Q7: Consider a data frame containing data about students i.e. name, gender and passing division:

```python
import numpy as np
import pandas as pd

df = pd.read_csv('F:\CS\sem 5\Data Analysis and Visualisation\Practicals\Student_7.csv')
df
```

## a. Perform one hot encoding of the last two columns of categorical data using the get_dummies() function.

```python
df1 = pd.get_dummies(df, columns=['Gender', 'Pass_Division'])
df1
```

## b. Sort this data frame on the “Birth Month” column (i.e. January to December). Hint: Convert Month to Categorical.

```python
months = ["January", "February", "March", "April", "May", "June", "July",
          "August", "September", "October", "November", "December"]

df1['Birth_Month'] = pd.Categorical(df1['Birth_Month'], categories=months, ordered=True)
df1.sort_values(by='Birth_Month', inplace=True)
df1
```
